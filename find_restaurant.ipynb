```python
import pandas as pd
```

```python
data = pd.read_json('amenities-vancouver.json.gz', lines = True)
```

```python
data
```

```python
food = data[(data.amenity == 'fast_food' )| (data.amenity == 'restaurant')]
```

```python
food
```

```python
food = food.groupby('name').count()
```

```python
food['chain'] = (food.tags >=2)
```

```python
food = food.sort_values(by='tags', ascending=False)
food = food.reset_index()
food
```

```python
food = food[['name','chain']]
food
```

```python
food = pd.merge(data, food, on = 'name')
```

```python
food = food[['lat', 'lon', 'name', 'chain']]
```

```python
food.to_csv("restaurant.csv", index = False)
```
